```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/zomato.csv', engine= 'python', on_bad_lines='skip')

df.head()
```

```python
print(df.shape)
df.isna().sum()
```

```python
df = df.dropna()
df.isna().sum()
```

```python
print(df.columns)
unwanted_cols = ['menu_item','phone']

df= df.drop(columns=unwanted_cols)

df.head(2)
```

```python
df.loc[1,'dish_liked']
```

```python
df['total address'] = df['address'] + ', ' +df['location'] + ', ' +  df['listed_in(city)']
df['food items'] = df['dish_liked'] + ', ' +df['cuisines']
original_df = df.copy()

df = df[['listed_in(type)','food items', 'rest_type','total address','book_table','online_order', 'rate', 'approx_cost(for two people)']]
df.head(2)
```

```python
df.to_csv('cleaned_df.csv')
```

```python
from nltk.tag.brill import Word
import re
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

nltk.download('stopwords')
nltk.download('wordnet')

stop_words = set(stopwords.words('english'))
lematizer = WordNetLemmatizer()

def preprocessing_data (text) -> str:

    text = str(text)
    # Converting the text into lower case
    text = text.lower()

    # Removing punctuvations
    text = re.sub(r'[^a-zA-Z0-9\s]','',text)

    # Removing stop words
    words = [w for w in text.split() if w not in stop_words]

    # Lemotization
    words = [lematizer.lemmatize(w) for w in words]

    return ' '.join(words)



```

```python
df['total address'] = ' - ' + df['total address']
df['total address'] = df['total address'].apply(preprocessing_data)
df.head(2)
```

```python
df.loc[df['rate'] == 'NEW','rate'] = df['rate'].mode()[0]
df['rate'].unique()
```

```python
df['rate'] = df['rate'].apply(lambda x: x.replace('/5','')).str.strip().astype('float')
df['book_table'] = df['book_table'].map({'Yes': 1, 'No': 0}).astype('int')
df['online_order'] = df['online_order'].map({'Yes': 1, 'No': 0}).astype('int')
df.info()
```

```python
df.head(2)
```

```python
df['rest_type'] = df['rest_type'].apply(preprocessing_data)
df['food items'] = df['food items'].apply(preprocessing_data)

df.head(2)
```

```python
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].apply(lambda x: x.replace(',', '')).astype(float)
```

```python
df['listed_in(type)'] = df['listed_in(type)'].replace('Drinks & nightlife' , 'Drinks and Nightlife')
df['listed_in(type)'].unique()
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
```

```python
df['listed_in(type)'] = label_encoder.fit_transform(df['listed_in(type)'])
df.head(2)
```

```python
df['final_phrase'] = df['food items'] + ' ' +df['rest_type'] + ' ' + df['total address']

df.head(2)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features= 1000)

tfidf_matrix = tfidf.fit_transform(df['final_phrase']).toarray()
```

```python
from sklearn.preprocessing import StandardScaler

four_scaler = StandardScaler()
one_scaler = StandardScaler()

cost_matrix = four_scaler.fit_transform(df.drop(columns= ['rest_type','rate', 'total address','food items','final_phrase']))

rate_array = one_scaler.fit_transform(df['rate'].values.reshape(-1, 1))
```

```python
final_matrix = np.hstack((tfidf_matrix, cost_matrix))

final_matrix
```

```python
df.loc[0:3, ['final_phrase','listed_in(type)', 'book_table', 'online_order', 'approx_cost(for two people)']]
```

```python
final_matrix[0:3]
```

```python
rate_array
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
network = keras.Sequential([
    layers.Dense(256, activation='relu',input_dim=final_matrix.shape[1]),
    layers.Dropout(0.3),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='swish'),
    layers.Dense(1, activation='linear')
])
```

```python
network.compile(optimizer='adam', loss='mean_squared_error', metrics= ['mae'])
```

```python
history =network.fit(final_matrix, rate_array, epochs= 150, batch_size=400, validation_split=0.2, verbose= 1)
```

```python
network.save('network.keras')
```
